```python
#import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#importing dataset 
df=pd.read_csv('/Users/abhisheksinha/Desktop/movies.csv')
df.head()
```

```python
#No. of rows and columns in the dataset
df.shape
```

```python
#looking at the data types of each variable 
df.info()
```

```python
#checking for missing values
df.isnull().sum()
```

```python
df.describe(include='all')
```

```python
#-replacing missing values in this budget with its median value
##median value for budget
df['budget'].median()
#although from logical point of view it doesn't make sense to replace these values b/c each of these movies have different budgets
```

```python
df['budget'].fillna(20500000.0, inplace=True)
```

```python
#since rest of the columns have less than 5% of the missing values therefore those missing vales were removed
df.dropna(subset=['gross','rating','released','score','votes','writer','star','company','country','runtime'],axis=0,inplace=True)
```

```python
df['budget']=df['budget'].astype('int')
df['gross']=df['gross'].astype('int')
```

```python
#To check movie with highest gross income
#pd.set_option('display.max_rows',None)
df.sort_values(by=['gross'], inplace=False,ascending=False).head()
```

```python
#Looking for duplicate values
df['company'].duplicated(keep='first').head()
```

```python
#Looking at the correlation between budget and gross earnings
fig,ax=plt.subplots(figsize=(10,8))
sns.scatterplot(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross earnings')
#plot shows that there is a positive correlation
```

```python
fig,ax=plt.subplots(figsize=(10,8))
sns.regplot(x=df['budget'],y=df['gross'],scatter_kws={"color":"red"},line_kws={"color":"blue"})
```

```python
#finding out the actual correlation between gross and budget
df[['gross','budget']].corr()
```

```python
#finding out the actual correlation for all the numerical variables in the dataset
df.corr()
```

```python
fig,ax=plt.subplots(figsize=(10,8))
sns.heatmap(df.corr(),annot=True)
plt.title('Correlation Matrix for Numeric Variables')
plt.xlabel('Movie variables')
plt.ylabel('Movie variables')
```

```python
#In order to look for the correlation for categorical variables as well, 
#we need to convert the string variables into categories(all the distinct values are numbered (dummy variables))
df_numerized=df
for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype=='object'):
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes
df_numerized.head()
        
        
        
```

```python
fig,ax=plt.subplots(figsize=(15,10))
sns.heatmap(df_numerized.corr(),annot=True)
plt.title('Correlation Matrix for Numeric Variables')
plt.xlabel('Movie variables')
plt.ylabel('Movie variables')
```

```python
df_numerized.corr()
```

```python
#this will sort the values of correlation for each variable in descending order
corr_matrix=df_numerized.corr()
corr_pairs=corr_matrix.unstack()
corr_pairs.head()
```

```python
sorted_pairs=corr_pairs.sort_values()
sorted_pairs
```

```python
high_correlation=sorted_pairs[(sorted_pairs)>0.5]

high_correlation[high_correlation!=1.0]
```
